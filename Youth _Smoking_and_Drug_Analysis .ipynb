# Univariate Analysis

```python
# Univariate Analysis with Bar Labels
numerical_features = df.select_dtypes(include=['float64', 'int64']).columns

for feature in numerical_features:
    plt.figure(figsize=(16, 3))
    ax = sns.histplot(df[feature], bins=30, kde=True)
    plt.title(f'Distribution of {feature}')
    
    # Adding bar labels
    for container in ax.containers:
        ax.bar_label(container, fmt='%.0f')  # Format the labels as needed
    
    plt.show()
```

# What is the distribution of smoking prevalence among youth?

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(18, 5))
ax = sns.histplot(df['Smoking_Prevalence'], bins=30, kde=True)
plt.title('Distribution of Smoking Prevalence')
plt.xlabel('Smoking Prevalence')
plt.ylabel('Frequency')

# Adding bar labels
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')

plt.show()
```

# What are the trends in smoking prevalence among youth over the years?

```python

plt.figure(figsize=(15, 6))
ax = sns.lineplot(data=df, x='Year', y='Smoking_Prevalence', ci=None)

# Add labels above each point
for line in ax.lines:
    for x, y in zip(line.get_xdata(), line.get_ydata()):
        ax.text(x, y, f'{y:.1f}%', ha='center', va='bottom')  # Formatting the label

plt.title('Smoking Prevalence Trends Over Years')
plt.ylabel('Smoking Prevalence Percentage')
plt.show()
```

# What are the trends in drug experimentation over the years, and how do the values compare year by year?

```python

plt.figure(figsize=(16, 6))
ax = sns.lineplot(data=df, x='Year', y='Drug_Experimentation', ci=None)

# Add labels above each point
for line in ax.lines:
    for x, y in zip(line.get_xdata(), line.get_ydata()):
        ax.text(x, y, f'{y:.1f}%', ha='center', va='bottom')  # Formatting the label

plt.title('Drug Experimentation Trends Over Years')
plt.ylabel('Drug Experimentation Percentage')
plt.show()
```
